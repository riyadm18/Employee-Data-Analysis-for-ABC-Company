```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("abc co.csv")
df.head()
```

```python
# Step 2.1: Replace the 'Height' column with random values between 150 and 180 (in cm)
np.random.seed(42)
df['Height'] = np.random.randint(150, 181, size=len(df))

# Step 2.2: Convert columns to appropriate data types
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Weight'] = pd.to_numeric(df['Weight'], errors='coerce')

# Step 2.3: Preview cleaned data
df.head()
```

```python
# Distribution and percentage of employees per team
team_counts = df['Team'].value_counts()
team_percent = (team_counts / len(df)) * 100

# Visualization
plt.figure(figsize=(10, 6))
sns.barplot(x=team_counts.values, y=team_counts.index, palette="viridis")
plt.title("Team-wise Employee Distribution")
plt.xlabel("Number of Employees")
plt.ylabel("Team")
plt.tight_layout()
plt.show()

# Table
team_distribution = pd.DataFrame({
    "Team": team_counts.index,
    "Employee Count": team_counts.values,
    "Percentage (%)": team_percent.values.round(2)
})
team_distribution
```

```python
position_counts = df['Position'].value_counts()

# Visualization
plt.figure(figsize=(8,5))
sns.barplot(x=position_counts.index, y=position_counts.values, palette='Set2')
plt.title("Employee Count by Position")
plt.xlabel("Position")
plt.ylabel("Number of Employees")
plt.tight_layout()
plt.show()

position_counts_df = position_counts.reset_index()
position_counts_df.columns = ['Position', 'Count']
position_counts_df
```

```python
# Define age bins
bins = [0, 24, 29, 34, 100]
labels = ['<25', '25-29', '30-34', '35+']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

age_group_counts = df['Age Group'].value_counts().sort_index()

# Visualization
plt.figure(figsize=(7,5))
sns.barplot(x=age_group_counts.index, y=age_group_counts.values, palette='coolwarm')
plt.title("Distribution by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Employees")
plt.tight_layout()
plt.show()

age_group_counts
```

```python
# Group by Team
team_salary = df.groupby('Team')['Salary'].sum().sort_values(ascending=False)

# Group by Position
position_salary = df.groupby('Position')['Salary'].sum().sort_values(ascending=False)

# Plot Team Salary
plt.figure(figsize=(10, 5))
sns.barplot(x=team_salary.values, y=team_salary.index, palette='magma')
plt.title("Total Salary Expenditure by Team")
plt.xlabel("Total Salary")
plt.ylabel("Team")
plt.tight_layout()
plt.show()

# Plot Position Salary
plt.figure(figsize=(7, 5))
sns.barplot(x=position_salary.values, y=position_salary.index, palette='cubehelix')
plt.title("Total Salary Expenditure by Position")
plt.xlabel("Total Salary")
plt.ylabel("Position")
plt.tight_layout()
plt.show()

team_salary.head(1), position_salary.head(1)
```

```python
# Remove missing values for correlation plot
cor_df = df[['Age', 'Salary']].dropna()

# Scatter plot with regression line
plt.figure(figsize=(8, 6))
sns.regplot(x='Age', y='Salary', data=cor_df, scatter_kws={'alpha':0.6})
plt.title("Correlation Between Age and Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.tight_layout()
plt.show()

# Correlation coefficient
cor_df.corr()
```
